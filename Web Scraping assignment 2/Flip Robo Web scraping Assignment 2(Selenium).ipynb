```python
#importing all the necessary libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")   #connecting the webdriver 
```

```python
driver.get("https://www.naukri.com/")  #automated opening of www.naukri.com
```

```python
#Inserting 'Data Analyst' as designation
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys('Data Analyst')
```

```python
#Inserting Bangalore as location
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys('Bangalore')
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[6]")
search.click()  #click search button
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#scraping job titles
for i in driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]'):
    job_title.append(i.text)
job_title=job_title[0:10]
job_title
```

```python
#scraping locations for those jobs
for i in driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]'):
    job_location.append(i.text)
job_location=job_location[0:10]
job_location
```

```python
#scraping names of those companies
for i in driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]'):
    company_name.append(i.text)
company_name=company_name[0:10]
company_name
```

```python
#scraping experience required for those jobs
for i in driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]'):
    experience_required.append(i.text)
experience_required=experience_required[0:10]
experience_required
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))  #checking the length
```

```python
#creating a data frame
df1=pd.DataFrame({"Job Title":job_title,"Location":job_location,"Company":company_name,"Experience":experience_required})
df1
```
